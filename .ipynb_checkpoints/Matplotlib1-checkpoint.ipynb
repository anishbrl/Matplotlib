```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
```

```python
plt.plot(x, y)
```

```python
z=np.cos(x)
plt.plot(x,z)
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('Sine Wave')
plt.grid(True)
plt.show()
```

    For colors, you can use named colors like 'red', 'green', 'blue', or you can use hexadecimal color codes like '#FF5733'.
    For linestyles, you can use '-' for solid, '--' for dashed, '-.' for dash-dot, and ':' for dotted.
    For markers, you can use 'o' for circles, 's' for squares, '+' for plus signs, and many more.

```python
plt.figure(figsize=(8, 6)) 
plt.plot(x, y, color='green', linestyle='-', linewidth=4, marker='s', markersize=5, label='sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Wave')
plt.legend()
plt.grid(True)
plt.show()
```

plt.subplot(2, 1, 1) is a function call that creates a subplot grid with 2 rows and 1 column, and 
then selects the first subplot (the one at the top). 

```python
w=np.tan(x)
plt.figure(figsize=(8, 10)) 
plt.subplot(3, 1, 1)
plt.plot(x, y)
plt.title('Sin(x)')

plt.subplot(3, 1, 2)
plt.plot(x, z)
plt.title('Cos(x)')
plt.subplot(3, 1, 3)
plt.plot(x, w)
plt.title('tan(x)')
plt.tight_layout()  # Adjust spacing
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [20, 35, 30, 25]

# Plot
plt.bar(categories, values)

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D']
values = [20, 35, 30, 25]
plt.bar(categories, values, color='skyblue', edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
sizes = [25, 30, 20, 25]  # Sizes of each wedge
labels = ['A', 'B', 'C', 'D']  # Labels for each wedge

# Plot
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

# Title
plt.title('Pie Chart Example')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
sizes = [25, 30, 20, 25]
labels = ['A', 'B', 'C', 'D']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
#explode = (0, 0.1, 0, 0)  
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.title('Pie Chart Example')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
sizes = [25, 30, 20, 25]
labels = ['A', 'B', 'C', 'D']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0.1, 0, 0)  
plt.pie(sizes, labels=labels, explode=explode,colors=colors, autopct='%1.1f%%', shadow=True)
plt.title('Pie Chart Example')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.randint(10, 100, size=100) 
plt.scatter(x, y, s=sizes, alpha=0.5) 
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot Example')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(1000)
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```
